```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Define the coordinates for neurons in each layer
input_layer = [(0, 0.5), (0, 0.25), (0, 0.75)]  # 3 neurons
hidden_layer = [(0.5, 0.5), (0.5, 0.3), (0.5, 0.7), (0.5, 0.1), (0.5, 0.9)]  # 5 neurons
output_layer = [(1, 0.5)]  # 1 neuron
```

```python
# Create figure and axis
fig, ax = plt.subplots()
```

```python
# Plot neurons
for neuron in input_layer:
    circle = plt.Circle(neuron, 0.05, color='blue', ec='black')
    ax.add_artist(circle)
```

```python
for neuron in hidden_layer:
    circle = plt.Circle(neuron, 0.05, color='green', ec='black')
    ax.add_artist(circle)
```

```python
for neuron in output_layer:
    circle = plt.Circle(neuron, 0.05, color='red', ec='black')
    ax.add_artist(circle)
```

```python
# Plot connections
for neuron1 in input_layer:
    for neuron2 in hidden_layer:
        plt.plot([neuron1[0], neuron2[0]], [neuron1[1], neuron2[1]], color='black')
```

```python
for neuron1 in hidden_layer:
    for neuron2 in output_layer:
        plt.plot([neuron1[0], neuron2[0]], [neuron1[1], neuron2[1]], color='black')
```

```python
# Set axis limits and remove axis ticks
ax.set_xlim([-0.1, 1.1])
ax.set_ylim([-0.1, 1.1])
ax.set_xticks([])
ax.set_yticks([])
```

```python
# Show plot
plt.title('Simple Artificial Neural Network')
plt.show()
```
